```python
import os
os.getcwd()
import requests
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd





driver_path = './chromedriver'
driver = webdriver.Chrome(driver_path)
driver.implicitly_wait(2)

page_path = 'http://fifaonline4.inven.co.kr/dataninfo/rate/index.php?pg={}'  #1~17496page까지 있음
driver.get(page_path.format('1'))  #첫 페이지 실행


url = page_path.format('1')
res = requests.get(url)     #현재 페이지에 요청
soup = bs(res.text)         #현재 페이지 soup 정보 저장

i=1
while i:
    
    infos = soup.find_all('span', {'fifa4 text_area'})

    for info in infos:

#         url = page_path.format(i)   #현재 페이지주소 url에 저장
#         res = requests.get(url)     #현재 페이지에 요청
#         soup = bs(res.text)         #현재 페이지 soup 정보 저장

        infos = soup.find_all('span', {'fifa4 text_area'})

        #선수 정보 저장
        info_name = info.find('span', {'fifa4 name'}).get_text()
        info_position = info.find('span', {'fifa4 position'}).get_text()
        info_rate = info.find('span', {'fifa4 rate'}).get_text()
        info_position = info.find('span', {'fifa4 position'}).find('img').get('src')
        info_comment = info.find('span', {'fifa4 comment'}).get_text()

        #선수 정보 출력
#         print('선수명 :', info_name)
#         print('능력치 :', info_position)
#         print('평점 :', info_rate)
#         print('선호포지션 :', info_position)
#         print('comment :', info_comment)
#         print()

        with open('fifa_inven_info.txt', 'a', -1, 'utf-8') as file:    #데이터 txt파일에 저장
            file.write(info_name + '\t')
            file.write(info_position + '\t')
            file.write(info_rate + '\t')
            file.write(info_position + '\t')
            file.write(info_comment +'\n')
    
   
    try:
        driver.find_element_by_xpath('//*[@id="fifaonline4Db"]/div[3]/span/span[16]/a').click() #'다음' 페이지로 이동
        i+=1           #다음페이지 url 가져오기위해 i에 1을 더함
        url = page_path.format(i)   #현재 페이지주소 url에 저장
        res = requests.get(url)     #현재 페이지에 요청
        soup = bs(res.text)         #현재 페이지 soup 정보 저장
        
        
    except:
        
        driver.find_element_by_xpath('//*[@id="fifaonline4Db"]/div[3]/span/span[14]/a').click() #'다음' 페이지로 이동
        time.sleep(3)

        i+=1           #다음페이지 url 가져오기위해 i에 1을 더함
        url = page_path.format(i)   #현재 페이지주소 url에 저장
        res = requests.get(url)     #현재 페이지에 요청
        soup = bs(res.text)         #현재 페이지 soup 정보 저장
        
    print('page: ' , i, ' 진행중')


time.sleep(3)

driver.quit()
```
